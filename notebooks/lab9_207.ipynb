```python
# ACT1-Import the appropriate libraries
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Construct pandas DataFrame
data = {
    'Weather': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Sunny',
                'Overcast', 'Rain', 'Rain', 'Overcast', 'Sunny'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool',
                    'Cool', 'Mild', 'Mild', 'Hot', 'Hot'],
    'Play': ['Yes', 'Yes', 'No', 'Yes', 'No',
             'No', 'Yes', 'Yes', 'No', 'No']
}
df = pd.DataFrame(data)

# Generate list of features and label
features = df[['Weather', 'Temperature']].copy()  # Use .copy() to avoid warning
labels = df['Play']

# Encode categorical features and label
le_weather = LabelEncoder()
le_temp = LabelEncoder()
le_play = LabelEncoder()

features.loc[:, 'Weather'] = le_weather.fit_transform(features['Weather'])
features.loc[:, 'Temperature'] = le_temp.fit_transform(features['Temperature'])
labels = le_play.fit_transform(labels)

# Call the chosen classifier container
classifier = DecisionTreeClassifier()

# Train the model using features and label
classifier = classifier.fit(features, labels)

# Include a new entry: Weather = Overcast, Temperature = Mild
new_data = pd.DataFrame({
    'Weather': le_weather.transform(['Overcast']),
    'Temperature': le_temp.transform(['Mild'])
})

# Predict the class of the new entry
prediction = classifier.predict(new_data)

# Decode and print prediction
print("Predicted:", le_play.inverse_transform(prediction)[0])
```

```python
#Activity 2
# Import the appropriate libraries, Pandas, numpy and sklearn using
import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Construct panda data frame
df = pd.read_csv("covid_related_disease_data.csv")

# Generate the list of features
features_list = ['Age', 'Gender', 'Preexisting_Condition', 'Vaccination_Status', 'Doses_Received', 'BMI']
df = df[features_list + ['Severity']].dropna()

# Generate the list of labels. (Note: Number of labels must be same as the number of feature rows)
le_gender = LabelEncoder()
le_condition = LabelEncoder()
le_vaccine = LabelEncoder()
le_severity = LabelEncoder()

df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Preexisting_Condition'] = le_condition.fit_transform(df['Preexisting_Condition'])
df['Vaccination_Status'] = le_vaccine.fit_transform(df['Vaccination_Status'])
df['Severity'] = le_severity.fit_transform(df['Severity'])

features_1 = df[features_list]
labels = df['Severity']

# Call the chosen classifier container that will train the feature data frame
Classifier = DecisionTreeClassifier()

# Train the model using features and label
Classifier = Classifier.fit(features_1, labels)

# Include a new entry having only features
new_entry = pd.DataFrame([{
    'Age': 34,
    'Gender': le_gender.transform(['Female'])[0],
    'Preexisting_Condition': le_condition.transform(['Hypertension'])[0],
    'Vaccination_Status': le_vaccine.transform(['Yes'])[0],
    'Doses_Received': 3,
    'BMI': 23.5
}])

# Predict the class of the new entry using
prediction = Classifier.predict(new_entry)
print("Predicted:", le_severity.inverse_transform(prediction)[0])
```

```python
#Act3
# Import the required libraries.
from sklearn import preprocessing 
from sklearn.naive_bayes import GaussianNB 

# Construct the data (Table 9.2)
outlook = ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Overcast', 'Sunny']
temperature = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Mild', 'Mild', 'Hot', 'Hot']
humidity = ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal']
wind = ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Strong']
play = ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No']

# Assign features and label variables.
le_outlook = preprocessing.LabelEncoder()
le_temp = preprocessing.LabelEncoder()
le_humidity = preprocessing.LabelEncoder()
le_wind = preprocessing.LabelEncoder()
le_play = preprocessing.LabelEncoder()

# Converting string into numbers
outlook_encoded = le_outlook.fit_transform(outlook)
temperature_encoded = le_temp.fit_transform(temperature)
humidity_encoded = le_humidity.fit_transform(humidity)
wind_encoded = le_wind.fit_transform(wind)
label = le_play.fit_transform(play)

# Combine all the features in a single variable (list of tuples)
features = list(zip(outlook_encoded, temperature_encoded, humidity_encoded, wind_encoded))

# Generate a model using naive bayes classifier
# 1. Create naïve bayes classifier.
model = GaussianNB()

# 2. Fit the dataset on classifier.
model.fit(features, label)

# 3. Perform prediction.
# Test: Outlook=Rain, Temperature=Cool, Humidity=High, Wind=Strong
test_data = [[
    le_outlook.transform(['Rain'])[0],
    le_temp.transform(['Cool'])[0],
    le_humidity.transform(['High'])[0],
    le_wind.transform(['Strong'])[0]
]]

predicted = model.predict(test_data)
print("Predicted:", le_play.inverse_transform(predicted)[0])
```

```python
# Import the required libraries.
from sklearn import preprocessing 
from sklearn.naive_bayes import GaussianNB 
import pandas as pd

# Load the dataset
df = pd.read_csv("covid_related_disease_data.csv")

# Assign features and label variables.
features_list = ['Age', 'Gender', 'Preexisting_Condition', 'Vaccination_Status', 'Doses_Received', 'BMI']
label_column = 'Severity'

df = df[features_list + [label_column]].dropna()

# Perform label encoding on all columns
le_gender = preprocessing.LabelEncoder()
le_condition = preprocessing.LabelEncoder()
le_vaccine = preprocessing.LabelEncoder()
le_label = preprocessing.LabelEncoder()

df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Preexisting_Condition'] = le_condition.fit_transform(df['Preexisting_Condition'])
df['Vaccination_Status'] = le_vaccine.fit_transform(df['Vaccination_Status'])
df['Severity'] = le_label.fit_transform(df['Severity'])

# Creating individual feature variables
age_encoded = df['Age']
gender_encoded = df['Gender']
condition_encoded = df['Preexisting_Condition']
vaccine_encoded = df['Vaccination_Status']
doses_encoded = df['Doses_Received']
bmi_encoded = df['BMI']
label = df['Severity']

# Combine all the features in a single variable (list of tuples)
features = list(zip(age_encoded, gender_encoded, condition_encoded, vaccine_encoded, doses_encoded, bmi_encoded))

# Generate a model using naive bayes classifier
# 1. Create naïve bayes classifier.
model = GaussianNB()

# 2. Fit the dataset on classifier.
model.fit(features, label)

# 3. Perform prediction.
# New test data (example): Age=40, Gender=Male, Condition=Asthma, Vaccinated, 2 doses, BMI=24.5
test_data = [[
    40,
    le_gender.transform(['Male'])[0],
    le_condition.transform(['Asthma'])[0],
    le_vaccine.transform(['Yes'])[0],
    2,
    24.5
]]

predicted = model.predict(test_data)
print("Predicted:", le_label.inverse_transform(predicted)[0])
```
