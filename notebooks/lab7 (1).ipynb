```python
import numpy as np
import matplotlib.pyplot as plt

# Data from Table A-11
date = ['10-03-22', '10-04-22', '10-05-22', '10-06-22', '10-07-22']
open_val = np.array([774.25, 776.030029, 779.309998, 779, 779.659973])
high = np.array([776.065002, 778.710022, 782.070007, 780.47998, 779.659973])
low = np.array([769.5, 772.890015, 775.650024, 775.539978, 770.75])
close = np.array([772.559998, 776.429993, 776.469971, 776.859985, 775.080017])

# Plotting 4 pie charts for each financial attribute
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].pie(open_val, labels=date, autopct='%1.1f%%', startangle=140)
axs[0, 0].set_title('Open Prices Distribution')

axs[0, 1].pie(high, labels=date, autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('High Prices Distribution')

axs[1, 0].pie(low, labels=date, autopct='%1.1f%%', startangle=140)
axs[1, 0].set_title('Low Prices Distribution')

axs[1, 1].pie(close, labels=date, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Close Prices Distribution')

# Adjust layout and display
plt.suptitle('Financial Data Distribution by Date (Oct 3–7, 2022)', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
#Take 2D and 3D array and perform transpose function. 

# 2D array
np_2d = np.array([[1, 2, 3],
                  [4, 5, 6]])

# Transpose of 2D array (rows become columns)
transpose_2d = np_2d.T

# 3D array (2 matrices of 2x2)
np_3d = np.array([[[1, 2],
                   [3, 4]],
                  
                  [[5, 6],
                   [7, 8]]])

# Transpose of 3D array: let's swap axes 0 and 1
transpose_3d = np.transpose(np_3d, (1, 0, 2))

# Print results
print("Original 2D array:\n", np_2d)
print("Transposed 2D array:\n", transpose_2d)

print("\nOriginal 3D array:\n", np_3d)
print("Transposed 3D array (axes 0 and 1 swapped):\n", transpose_3d)
```

```python
#multiply a 5x3 matrix by a 3x2 matrix create a real matrix product using dot ( ). 
def dot_product():
    # 5x3 matrix
    mat1 = np.array([
        [2, 3, 4],
        [1, 8, 9],
        [4, 7, 8],
        [5, 9, 0],
        [3, 5, 1]
    ])

    # 3x2 matrix
    mat2 = np.array([
        [3, 2],
        [5, 6],
        [9, 0]
    ])

    # Matrix multiplication using dot
    result = np.dot(mat1, mat2)

    print("Result of dot product:\n", result)

# Call the function
dot_product()
```

```python
#Write a Python program to create a 2-D array with reshape (5, 5) and filled 1 on the border and 0 inside using slicing method.

# Step 1: create a 5x5 array full of 1s
arr = np.ones((5, 5), dtype=int)

# Step 2: use slicing to set the center part to 0
arr[1:-1, 1:-1] = 0

# Step 3: print the final result
print(arr)
```

```python
#Write a Python program to create a 3-D array with ones on the diagonal and zeros elsewhere

# Create a 3D array filled with zeros
arr3d = np.zeros((3, 3, 3), dtype=int)

# Fill the diagonal positions (i == j == k) with 1
for i in range(3):
    arr3d[i] = np.eye(3, dtype=int)[i]  # assign each 2D slice with eye row

# Print the final array
print(arr3d)
```

```python
#Write a Python program to find the real and imaginary parts of an array of complex numbers.

# Create an array with complex numbers
arr = np.array([2 + 3j, 4 - 5j, 6 + 0j])

# Print the full array
print("Complex array:", arr)

# Print real parts
print("Real parts:", arr.real)

# Print imaginary parts
print("Imaginary parts:", arr.imag)

```
